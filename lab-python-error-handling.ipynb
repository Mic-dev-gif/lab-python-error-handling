````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                # Prompt for quantity and check if it's valid
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    # Raise error if quantity is negative
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                # Handle invalid input (non-numeric or negative)
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
````

````python
def calculate_total_price(customer_orders):
    total_price = 0
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                # Prompt for price
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    # Raise error if price is negative
                    raise ValueError("Invalid price! Please enter a non-negative price.")
                total_price += price
                valid_price = True
            except ValueError as error:
                # Handle invalid input (non-numeric or negative)
                print(f"Error: {error}")
    return total_price
````

````python
def get_customer_orders(inventory):
    customer_orders = set()  # Using a set to ensure unique products are ordered
    valid_orders = False
    while not valid_orders:
        try:
            # Ask for the number of customer orders
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders < 0:
                raise ValueError("Number of orders cannot be negative. Please enter a valid number.")
            valid_orders = True
        except ValueError as error:
            print(f"Error: {error}")

    # Collect the customer orders
    for _ in range(num_orders):
        valid_product = False
        while not valid_product:
            product_name = input("Enter the name of a product that a customer wants to order: ").lower()
            if product_name in inventory and inventory[product_name] > 0:
                customer_orders.add(product_name)
                valid_product = True
            else:
                print(f"Error: '{product_name}' is not available or out of stock. Please enter a valid product.")
                
    return customer_orders
````

````python
def main():
    # Sample product list
    products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']
    
    # Step 1: Initialize the inventory
    inventory = initialize_inventory(products)
    
    # Step 2: Get customer orders
    customer_orders = get_customer_orders(inventory)
    
    # Step 3: Calculate total price of customer orders
    total_price = calculate_total_price(customer_orders)
    
    # Output order details
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {len(customer_orders)}")
    print(f"Percentage of Unique Products Ordered: {100 * len(customer_orders) / len(products):.2f}%")
    
    # Output updated inventory (after orders are processed)
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
    
    # Output total price
    print(f"\nTotal Price: {total_price:.2f}")

# Run the main function
main()
````
